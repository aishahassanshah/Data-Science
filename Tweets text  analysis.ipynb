```python
dic_kim={} #dictionaries are unordered
for word in word_list_kim:
    dic_kim[word]=dic_kim.get(word,0)+1
```

```python
word_freq_kim=[]
for key,value in dic_kim.items():
    word_freq_kim.append((value,key))
```

```python
word_freq_kim.sort(reverse=True)
print(word_freq_kim)
```

```python
#Log odd ratio calculation of each word by both users ,list and dictionary are already created just plugin in to log ratio formula.

import math
a=word_list_BO 
b=word_list_kim


dict_a = dic_BO
dict_b = dic_kim

s=list(set(a) & set(b))
a=sum(dict_a.values())
b=sum(dict_b.values())
log_ratio_BO=[]
log_ratio_kim=[]


for i in s :
#   print(i)

    log_ratio_BO.append(math.log2((dict_a[i]+1/a+1)/(dict_b[i]+1/b+1)))
    log_ratio_kim.append(math.log2((dict_b[i]+1/b+1)/(dict_a[i]+1/a+1)))



# print(log_ratio_BO)

Log2_BO=pd.DataFrame(log_ratio_BO)
Log2_BO.rename(columns={0:'Log odd ratio of Barack Obama'},inplace=True)
print(Log2_BO)                      
                        
                
Log2_kim=pd.DataFrame(log_ratio_kim)
Log2_kim.rename(columns={0:'Log odd ratio of Kim kardashsian'},inplace=True)
print(Log2_kim)
```
